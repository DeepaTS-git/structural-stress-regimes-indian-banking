```python

import pandas as pd
import warnings
# Skipping top metadata rows

npa_file = r"../data/Gross and Net NPAs of Scheduled Commercial Banks - Bank Group-Wise.xlsx"
# Load data, skip metadata
npa_df = pd.read_excel(npa_file, skiprows=11, header=None)

warnings.filterwarnings("ignore", category=UserWarning, module="openpyxl")

npa_df = npa_df.drop(npa_df.columns[0], axis=1)

# Assigning correct column names manually
npa_df.columns = [
    "Year",
    "Advances_Gross",
    "Advances_Net",
    "Gross_NPA_Amount",
    "Gross_NPA_pct_Gross_Adv",
    "Gross_NPA_pct_Total_Assets",
    "Net_NPA_Amount",
    "Net_NPA_pct_Net_Adv",
    "Net_NPA_pct_Total_Assets"
]
```
